## Fusão das bases de dados

```python
merged_df = pd.merge(titainic,portos,left_on='Embarked',right_on='City')
```

Considerando que cada port de embarque está em uma cidade diferente, cada cidade está em um país diferente e cada país fala uma língua diferente, temos que estas variáveis são redundantes. Assim, manteremos apenas o local de embarque. 

Além disso, PassengerId só representa a posição do registro na tabela. Portanto, não pertence diretamente ao problema e será removida.

```python
merged_df = merged_df.drop('PassengerId',axis=1)
merged_df = merged_df.drop('City',axis=1)
merged_df = merged_df.drop('Country',axis=1)
merged_df = merged_df.drop('Main Language',axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = merged_df.drop('Survived',axis=1)
y = merged_df['Survived']
```

```python
clfs = [DecisionTreeClassifier(),
        RandomForestClassifier(),
        RandomForestClassifier(criterion='entropy',max_depth=8),
        RandomForestClassifier(criterion='log_loss',max_depth=10),
        SVC(),
        KNeighborsClassifier()]
```

```python
y.info()
```

```python
from sklearn.model_selection import cross_val_score

classifiers = []
scores = []

for clf in clfs:
    cv_scores = cross_val_score(clf,X,y, cv=10, scoring='f1')
    for s in cv_scores:
        classifiers.append(str(clf))
        scores.append(float(s))
```

```python
results = pd.DataFrame({'classifier':classifiers,'F-score':scores})
```

```python
sns.set(rc={'figure.figsize':(25,8)})
sns.violinplot(data=results,x='classifier',y='F-score')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['gini','entropy','log_loss'],
              'max_depth':[2,3,4,5,6,7,8,9,10],
              }

grid = GridSearchCV(RandomForestClassifier(),param_grid,cv=10,scoring='f1')
```

```python
grid.fit(X,y)
grid.best_params_
```
